```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
def interest_paid(loan_amt, n_years, apr, payment):
    interest = 0
    loan_amt = loan_amt
    for i in range(0,n_years):
        end_balance = max(loan_amt - payment + (loan_amt * apr),0)
        interest += max((loan_amt * apr),0)
        loan_amt = max(end_balance,0)
    return interest

def loan_amortization(loan_amt, n_years, apr):
    payment = (loan_amt * apr) / (1 - (1 + apr) ** (-n_years))

    amort_sched = pd.DataFrame(columns=["Period", "Beginning Balance", "Payment", "Interest",
                                    "Principal Reduction", "Ending Balance"])
    amort_sched['Period'] = range(1,n_years + 1)
    amort_sched['Payment'] = np.around(payment, 4)
    for i in range(0,n_years):
        if i == 0:
            amort_sched['Beginning Balance'][i] = loan_amt
            amort_sched['Interest'] = amort_sched['Beginning Balance'] * apr
            amort_sched['Principal Reduction'] = amort_sched['Payment'] - amort_sched['Interest']
            amort_sched['Ending Balance'] = amort_sched['Beginning Balance'] - amort_sched['Principal Reduction']
        else:
            amort_sched['Beginning Balance'][i] = amort_sched['Ending Balance'][i-1]
            amort_sched['Interest'] = amort_sched['Beginning Balance'] * apr
            amort_sched['Principal Reduction'] = amort_sched['Payment'] - amort_sched['Interest']
            amort_sched['Ending Balance'] = amort_sched['Beginning Balance'] - amort_sched['Principal Reduction']
    amort_sched = amort_sched.astype(float).round(2)

    
    print("You will pay $" + str(round(amort_sched["Interest"].sum(), 2)) + " in interest on this loan.\nIncreasing the payment by 10% would decrease the interest paid to\n$" + str(round(interest_paid(loan_amt, n_years, apr, payment * 1.1),2))+", a savings of " + str(round(((amort_sched["Interest"].sum()/interest_paid(loan_amt, n_years, apr, payment * 1.1))-1)*100,2)) + "%.")
    return amort_sched.set_index('Period')
```

```python
loan_amt = 42850.00
n_years = 12
apr = .032

loan_amortization(loan_amt, n_years, apr)
```
